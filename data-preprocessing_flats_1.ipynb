```python
# Starting with column one by one
# Society (Categorical column)

df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
# Since some society have rating and some don't , so removing the rating part
# Also same society name in capital and small are counted as 2 different category, so making all of them in lower case through Regular Expression

import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
# Checking the price column
# Observation : Value over crore is written in Crore and below that its written in lakh -> so making a similar unit in all of them

df['price'].value_counts()

```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']  # removing this column since very few number of missing values are there
```

```python
df.head()
```

```python
# Changing lac into crore
#x = '50 Lac'
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
            
        else:
            return round(float(x[0]),2)
        
# x = x.split(' ')
# print(x)
#treat_price(x)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
# price per sq feet - just need the number
df['price_per_sqft'].value_counts
```

```python
# Observatin from above: Some missing value, some rupee symbol, some slash (/) and comma(,) symbol, also removing any spaces if present using strip -> so cleaning all of them one by one
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
